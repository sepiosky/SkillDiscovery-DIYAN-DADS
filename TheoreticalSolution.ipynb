### الف)

$$
I(s^{\prime};z|s) = H(z|s) - H(z|s, s^{\prime}) = H(s^{\prime}|s) - H(s^{\prime}|s,z)
$$

از دو تعبیر فرمول اطلاعات متقابل بالا می توان برداشت کرد که هدف این آبجکتیو پراکندگی و تنوع حداکثری اسکیل ها و انتقال های ایجاد شده در محیط (به منظور اکسپلوریشن مناسب) و همچنین قابل تشخیص بودن اسکیل ها با توجه به انتقال ها و بالعکس می باشد که این امر نیز باعث تنوع در اسکیل ها می شود. در فرم اول بیشینه کردن ترم اول باعث حداکثر کردن تنوع اسکیل ها در استیت های مختلف می شود و کمینه کردن ترم دوم باعث قابل شناسایی بودن اسکیل ها با توجه به ترنزیشن مشاهده شده است (که این اتفاق نیز به صورت ضمنی باعث تنوع در اسکیل ها می شود زیرا وقتی با توجه به ترنزیشن اسکیل شناسایی شود، تا جای ممکن سعی شده هیج ۲ اسکیلی ترنزیشن های مشابه تولید نکنند)

در فرم دوم نیز بیشینه کردن ترم اول باعث پراکندگی انتقال های بین استیت ها واکسپلوریشن مناسب می شود و کمینه کردن ترم دوم باعث قابل پییش بینی بودن انتقال ناشی از اعمال اسکیل ها و informative بودن اسکیل های یاد گرفته شده می شود.


### ب)


$$
I(s^\prime;z|s) = H(s^\prime|s) - H(s^\prime|s,z) = -\int p(s^\prime,s)log p(s^\prime|s)ds^\prime ds + \int p(s^\prime,s,z)logp(s^\prime|s,z)ds^\prime dsdz = \int p(s^\prime,s,z)log\frac{p(s^\prime|s,z)}{p(s^\prime|s)}ds^\prime dsdz = E_{s^\prime,s,z \sim p}[log\frac{p(s^\prime|s,z)}{p(s^\prime|s)}]
$$

عبارت $p(s^\prime|s,z) = \int p(s^\prime|s,a)\pi(a|s,z)da$ به دلیل نامشخص بودن داینامیک محیط قابل محاسبه نیست پس یک باند پایین وردشی برای آن بدست می آوریم:

$$
I(s^\prime;z|s) = E_{s^\prime,s,z \sim p}[log\frac{p(s^\prime|s,z)}{p(s^\prime|s)}] = E_{s^\prime,s,z \sim p}[log\frac{q_\phi(s^\prime|s,z)}{p(s^\prime|s)}] + E_{s,z \sim p} \big[E_{s^\prime \sim p(.|s,z)}[log p(s^\prime|s,z)-q_\phi(s^\prime|s,z)]\big] = \\ 
E_{s^\prime,s,z \sim p}[log\frac{q_\phi(s^\prime|s,z)}{p(s^\prime|s)}] + E_{s,z \sim p}[D_{KL}(p(s\prime|s,z),q_\phi(s\prime|s,z))] \\
\geq E_{s^\prime,s,z \sim p}[log\frac{q_\phi(s^\prime|s,z)}{p(s^\prime|s)}]
$$

در نتیجه فرآیند آموزش به ۲ بخش تقسیم می شود: ابتدا با سمپل گرفتن یک z برای اپیزود جاری، پالیسیی اسکیل کاندیشند در محیط اجرا می شود و تعدادی نمونه به شکل $(s,z,s^\prime)$ جمع آوری می شود.

سپس با این نمونه ها طبق رابطه ۵ مقاله گرادیان پارامتر های $\phi$ که همان درست نمایی تابع q روی سمپل های p است محاسبه می شود تا باند پایین با کمینه کردن فاصله توزیع وردشی و توزیع واقعی داینامیک اسکیل ها بیشینه شود.

بعد از بیشینه شدن باند پایین وردشی، باید آبجکتیو با توجه به توزیع وردشی بدست آمده بیشینه شود. همانطور که دیدیم باند پایین آبجکتیو به شکل  $E_{s^\prime,s,z \sim p}[log\frac{q_\phi(s^\prime|s,z)}{p(s^\prime|s)}]$ بدست آمد. این عبارت را می توان به صورت آبجکتیو یک ایجنت ری اینفورسمنت لرنینگ نگاه کرد که روی دنباله ای از $<(s,z), s^\prime>$ ها بهینه می شود که در این مساله z معمولا در طول اپیزود ثابت است (البته می تواند ثابت نباشد که باعث یافتن ترکیب های جالبی از اسکیل ها می شود اما پیچیدگی مساله را افزایش می دهد) می توان روی $<(s,z) , (s^\prime,z)>$ عمل کرد . از آنجایی که ورودی پالیسی هم هم استیت است هم اسکیل پس می توان استیت اسپیس را به شکل هایپر استیت (s,z) فرض کرد. داینامیک محیط هم همان داینامیک محیط اصلی است که با اضافه شدن اسکیل به استیت-امبدینگ میتوان آن را به شکل $P^\prime((s^\prime,z^\prime)|(s,z),a) = P(s^\prime|s,a) \delta(z^\prime-z)$ نوشت. اکشن اسپیس همان اکشن اسپیس محیط اصلی است اما قسمت مهم محیط جدید ریوارد فانکشن آن است که به صورت intrinsic و بدون توجه به ریوارد محیط و با توجه آبجکتیو بدست آمده از اطلاعات متقابل اسکیل به کار گرفته شده و ترنزیشن ایجاد شده در محیط به دست آمده است. با توجه به باند پایین آبجکتیو $E_{s^\prime,s,z \sim p}[log\frac{q_\phi(s^\prime|s,z)}{p(s^\prime|s)}]$ ریوارد محیط جدید برای هر ترنزیشن s به $s^\prime$ با بکار گرفتن اسکیل z برابر است با $log q_\phi(s^\prime|s,z) - log p(s^\prime|s)$ با توجه به مطالبی که در قسمت قبل گفته شد استفاده از این ریوارد باعث می شود که پالیسی برای اسکیل z اعمالی را در هر استیت انتخاب کند که از شاخصه های متمایز کننده اسکیل باشند (تابع discriminator به همین ویژگی پاداش می دهد و ترم دوم نیز به نوعی تعبیر نرمالیزیشن بین اسکور اسکیل های مختلف در ترنزیشن دیده شده را دارد اما نمی توان آن را حذف کرد زیرا نرمالیزشن در سطح استیت است و در طی مسیر scale ها ممکن است متفاوت باشد)



### ج)
ترم نرمالیزیشن ریوارد به شکل $p(s^\prime|s) = \int p(s^\prime|s,z)p(z|s)dz$ که دیدیم $p(s^\prime|s,z)$ قابل محاسبه نیست و همچنین $p(z|s) \propto p(s|z)p(z)$. برای جایگزین کردن این ترم در تابع پاداش با یک توزیع وردشی مشکل جابجا شدن باند پایین و بالا در بیشنیه سازی بوجود می آید:

$$
E_{s^\prime,s,z \sim p}[log q_\phi(s^\prime|s,z) - log p(s^\prime|s)] = E_{s^\prime,s,z \sim p}[log q_\phi(s^\prime|s,z) - log q(s^\prime|s) +log q(s^\prime|s) - log p(s^\prime|s)] = \\
= E_{s^\prime,s,z \sim p}[log q_\phi(s^\prime|s,z) - log q(s^\prime|s)] - E_s[D_{KL}(p(s^\prime|s) || q(s^\prime|s))] \leq E_{s^\prime,s,z \sim p}[log q_\phi(s^\prime|s,z) - log q(s^\prime|s)]
$$

همانطور که می بینیم با اضافه کردن توزیع وردشی جهت نامساوی عوض شد و بیشنیه کردن باند بالا نمیتواند برای بیشینه کردن پاداش ها استفاده شود (و واضحا برعکس آن هم باند پایین برای کمینه کردن منفی پاداش ها می دهد)

به جای استفاده از توزیع وردشی می توان در انتگرال به جای توزیع $p(s^\prime|s,z)$ از توزیع وردشی تخمین زننده آن که در قسمت قبل وارد کردیم استفاده کنیم، که با توجه به اینکه در قسمت آموزش این توزیع در چسباندن باند پایین مرحله قبل این توزیع تخمین زننده مناسبی است و استفاده از آن منطقی است. برای تخمین p(z|s) در انتگرال نیز می توان از یک توزیع وردشی q(z|s) استفاده کرد که البته همچنان باند بالا می دهد اما از آن صرفا برای تخمین p(z|s) با کمینه کردن فاصله کی ال آنها استفاده می شود. سپس انتگرال را به همراه این توزیع و توزیع وردشی اسکیل-داینامیک تخمین می زنیم. اما در مقاله ذکر شده است که در مثلا تسک لوکوموشن توزیع کاندیشنال z در استیت های مختلف باید نزدیک باشد پس به جای p(z|s) و یا توزیع وردشی آن از توزیع پرایور p(z) استفاده میکنیم که باعث می شود به دلیل سمپل گیری آسان از آن فرآیند آموزش و همچنین سمپل گرفتن از $p(s^\prime|s)$ سریع تر شود. یک توجیه دیگر نیز این است که ما میخواهیم اسکیل های یادگرفته شده از همه استیت ها قابل استفاده باشد و مختص استیت خاصی نباشند (تا حد ممکن اسکیل های  جنرال و اساسی باشند) \[که البته این توجیه بسیار به تسک بستگی دارد\].  پس در نهایت  $p(s^\prime|s) = \int p(s^\prime|s,z)p(z|s)dz$ با تخمین $p(s^\prime|s) = \int q_{\phi}(s^\prime|s,z)p(z)dz$ که برای تخمین آن نیز با سمپل گیری می توان L سمپل از توزیع پرایور z گرفت و انتگرال را به شکل $\frac{1}{L} \sum_{i=1}^{L} q_{\phi} (s^\prime|s,z_i)$ که $z_i \sim p(z)$ محاسبه کرد. 

در نهایت فرم کلی تخمین تابع پاداش به شکل زیر در می آید:

$$
r_z(s,a,s^\prime) = log \frac{q_\phi(s^\prime|s,z)}{\sum_{i=1}^{L} q_{\phi} (s^\prime|s,z_i)} + log L  \ \ \ \text{where} \ \ \ z_i \sim p(z)
$$

### د-۱)
همانطور که می دانیم در یادگیری بدون نظارت اسکیل ها، هدف یادگیری اسکیل هایی است که نمایانگر یک رفتار در غالب یک دنباله از اکشن های پیوسته و مرتبط به هم هستند. اگر این سطح انتزاع را در نظر بگیریم این اسکیل ها احتمال وقوع یک مسیر را در سطح بالاتری از اعمال و با توزیع $ p(s^\prime|s,z)$ مدل می کند که انگار با مارجینالایز کردن a در توزیع $\pi(s^\prime|s,z,a)$ اثر اعمال از بین رفته و اسکیل z تعیین کننده آینده ی مسیر است. حال با توجه به داشتن این توزیع هدف و بهینه کردن پالیسی $\pi$ برای نزدیک کردن هر چه بیشتر توزیع واقعی P به توزیع هدف N، اسکیل هایی مانند $\pi(.|., z)$ به نحوی اپتیمایز می شوند که به خوبی این انتزاع و پیوستگی زمانی در سطح اعمال را در قالب z های مختلف ایجاد کنند و در توزیع احتمال مسیر های مختلف در P مسیر هایی محتمل باشند که نمایانگر یک رفتار پیوسته و معنا دار در سطح اکشن ها باشند.


### د-۲)
در رابطه عبارت ۹، ترم سوم که $I(s_{t+1}|\{s_t,z\})$ است اطلاعات داینامیک محیط است که ثابت و نامشخص است و تاثیری در بهینه سازی پالیسی ندارد پس از آن صرف نظر می کنیم. همچنین ترم اول را به کمک قانون زنجیره ای اطلاعات متابل بازنویسی میکنیم تا به فرم معنادار تری برسیم. طبق قانون زنجیره ای انتروپی داریم: 

$$
I(A|\{B,C\}) = H(A) - H(A|B,C) = H(A) - H(A|B) + H(A|B) - H(A|B,C) = I(A;B) + I(A;C|B) \\
\\
\Rightarrow I(s_{t+1}|\{s_t,z\}) = I(s_{t+1};s_t) + I(s_{t+1};z|s_t) \\
\Rightarrow R(\pi) = \sum_{t=1}^{T-1} I(s_{t+1}|\{s_t,z\}) - \beta_t I(a_t|\{s_t,z\}) = \sum_{t=1}^{T-1} I(s_{t+1};s_t) + I(s_{t+1};z|s_t) - \beta_t I(a_t|\{s_t,z\}) \geq \sum_{t=1}^{T-1} I(s_{t+1};z|s_t) - \beta_t I(a_t|\{s_t,z\}) 
$$

باند پایین این آبجکتیو با توجیه استفاده از مدل هدف که در قسمت د-۱ ذکر شد کاملا مطابقت دارد: آبجکتیو پالیسی در مدل P به گونه ای است که با اپتیمایز کردن آن اطلاعات متقابل اسکیل و استیت بعدی در هر استیت بالا باشد و استیت بعدی به اسکیل وابسته باشد (که همان آبجکتیو قسمت الف نیز است که باعث پراکندگی و قابل پیش بینی بودن اسکیل ها می شد و همچنین در گراف هدف N، اتصال یال z به استیت های طول مسیر نمایانگر این موضوع است) و همچنین ترم دوم وجود اطلاعات متقابل بین [استیت و اسکیل فعلی] (هایپر استیت) و اکشن را مجازات می کند (که در گراف N با وجود نداشتن یال a به استیت بعدی وقتی z و استیت فعلی به استیت بعدی یال دارند هم معنی است). به تعبیری با دانستن اسکیل در هر استیت اکشن باید معلوم باشد و اطلاعات خاصی در متغیر تصادفی اکشن باقی نماند و استیت بعدی از اکشن مستقل باشد. در قسمت بعد ترم دوم این باند پایین همچنین استفاده از یک پالیسی بیشینه انتروپی مانند SAC را در بهینه سازی ابجکتیو اینفورمیشن تئوری قسمت الف توجیه می کند.


### د-۳)

$$
I(a_t;\{s_t,z\}) = E_{p(s_t,z,a_t)} [log \frac{\pi(a_t|s_t,z)}{\pi(a_t)}] = E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \frac{\pi(a_t|s_t,z)}{\pi(a_t)}]
$$

که چون محاسبه $\pi(a_t)$ که مارجینالایز شده $\pi(a_t|s_t,z)$ روی اسکیل ها و استیت های مختلف است قابل محاسبه نیست از یک باند پایین وردشی برای آن استفاده می کنیم:

$$
E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \frac{\pi(a_t|s_t,z)}{\pi(a_t)}] = E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \frac{\pi(a_t|s_t,z)}{p(a_t)}] - D_{KL}(\pi(a_t)||p(a_t)) \leq E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \frac{\pi(a_t|s_t,z)}{p(a_t)}]
$$

نامساوی از نامنفی بودن فاصله کی ال نتیجه شده است و توزیع $p(a_t)$ یک پرایور دلخواه روی اکشن ها است (مثلا یونیفورم). حال با جایگذاری این ترم در عبارت ۱۰ و حذف اکسپکتیشن روی $p(a_t)$ به دلیل ثابت بودن نسبت به پالیسی و بی تاثیر بودن در بهینه سازی به عبارت ۱۱ می رسیم که ترم اول ان همان آبجکتیو اینفورمیشن تئورتیک قسمت الف است و ترم دوم یک ترم بیشینه انتروپی است که از منظر مدل گرافی احتمالاتی و اینفورمیشن باتل نک استفاده از یک پالیسی بیشینه انتروپی در بهینه سازی بدون نظارت پیشنهاد شده در مقاله را توجیه کرد: 

$$
R(\pi) \geq \sum_{t=1}^{T-1} I(s_{t+1};z|s_t) - \beta_t E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \frac{\pi(a_t|s_t,z)}{p(a_t)}] \\
\text{optimizing lower-bound for $\pi$}\Rightarrow R(\pi) \geq \sum_{t=1}^{T-1} I(s_{t+1};z|s_t) - \beta_t E_{s_t, z \sim p, a_t \sim \pi(a_t|s_t,z)} [log \pi(a_t|s_t,z)]
$$

### ه)
در مقاله DIAYN اثرات اسکیل در توزیع های حالت پایا در نظر گرفته شده است. به همین منظور آبجکتیو ها نیز در این سطح بررسی شده اند. در DADS با بیشینه سازی $I(s_{t+1};z|s_t)$ سعی می شد که اطلاعات انتقال بین استیت ها از سطح اکشن ها بالاتر رفته و در اسکیل ها نهفته باشد. در DIAYN در نتیجه سعی می شود که اسکیل ها به جای کنترل ترنزیشن ها در سطح ترجکتوری، استیت هایی که ایجنت در آن ها قرار میگیرد را در سطح رفتار حالت پایا کنترل می کنند. به این منظور ترم $I(S;Z)$ در آبجکتیو بیشینه می شود که اطلاعات متقابل اسکیل و استیت های مشاهده شده توسط ایجنت را بالا می برد. همچنین مانند ترم دوم آبجکتیو در دیدگاه مدل های گرافی که سعی می کرد اطلاعات اکشن ها در ترنزیشن ها را با دانستن اسکیل کمینه کند، اینجا نیز برای اینکه تاکید شود اسکیل ها تعیین کننده توزیع پایای استیت ها هستند نه اکشن ها، با دانستن استیت اطلاعات متقابل اکشن ها و اسکیل را کمینه می کند یعنی $I(A;Z|S)-$ در آبجکتیو قرار می گیرد. همانطور که در ساده سازی آبجکتیو در مقاله DIAYN دیده می شود این پنالتی نیز سعی می کند که در سطح هایپر استیت ها یک سیاست بیشینه انتروپی را پیدا کند که با توجیه مدل های گرافی سازگار بود. اما این ترم دوم باعث ایجاد یک پنالتی در سطح استیت ها برای تصادفی بودن اکشن ها می شود (یعنی انتروپی میکسچر پالیسی اسکیل های مختلف در استیت ها را کاهش می دهد) که با اضافه کردن یک ترم انتروپی به آبجکتیو جبران شده است:

$$
F_{DIAYN}(\theta) = I(S;Z) + H(A|S) - I(A;Z|S) = (H(Z) - H(Z|S)) + H(A|S) - (H(A|S) - H(A|S,Z)) = H(Z) - H(Z|S) + H(A|S,Z)
$$

که تقریبا مشابه آبجکتیو DADS ترم اول پراکندگی اسکیل ها، ترم دوم قابل پیش بینی بودن اسکیل ها و ترم سوم بیشینه انتروبی بودن پالیسی را تشویق میکند.


تفاوت این دو تابع هدف اما در ترم دوم است. در مقاله DIAYN پراکندگی اسکیل ها تشویق شده اما در ترم دوم قابل پیش بینی بودن اسکیل ها در سطح استیت های مشاهده شده تشویق می شود. این اتفاق به نوعی باعث می شود که قابل پیش بینی بودن اسکیل ها با پراکندگی آن ها هم معنا شود زیرا فقط باید استیت های مختلف اسکیل های مختلف را به کار گیرند. این اتفاق باعث می شود که پالیسی و اسکیل ها در سطح ترنزیشن و اکشن ها قابل پیش بینی نباشند و واریانس بالایی در اسکیل ها و اکشن هایی که در آنها انجام می شود به وجود آید. اما در DADS با توجه به تابع هدف و نوع نگاه به آبجکتیو، علاوه بر پراکندگی اسکیل ها کنترل آنها نه تنها بر استیت های مشاهده شده بلکه بر داینامیک و انتقالات در محیط افزایش یابد. این اتفاق باعث می شود که اسکیل ها در سطح ترنزیشن ها و اکشن ها نیز قابل پیش بینی باشند (علاوه بر استیت های مشاهده شده) و در نتیجه واریانس رفتار اسکیل ها کاهش یابد و این کاهش واریانس می تواند پیوستگی زمانی و قابلیت ترکیب بهتر اسکیل ها را در مرحله پلنینگ و استفاده از اسکیل ها در حالت یادگیری نظارتی بهبود دهد


طبق هدفی که در DIAYN ذکر شد اطلاعات متقابل اسکیل ها و استیت های حالت پایا باید افزایش یابد و مانند مدل گرافی هدف قسمت قبلی اسکیل ها باید یک سطح انتزاع را روی اکشن ها ایجاد کنند. در نتیجه مدل گرافی واقعی و هدف در DIAYN به شکل زیر هستند:

مانند مقاله DADS و طبق عبارت ۳ صورت سوال، برای کمینه کردن فاصله ۲ توزیع به منظور بهینه سازی پالیسی ای که اسکیل های مناسب را با توجه به ساختار توزیع هدف یاد بگیرد به شکل زیر است:

$$
R(\pi) = -I_P + I_Q = -\sum_i I(P_i;Pa(P_i)) + \sum_i I(Q_i;Pa(Q_i)) = -I(a;\{s,z\}) + I(z;s) \\
 = -E_{s,a,z \sim p(s,a,z)}[log \frac{\pi(a|s,z)}{p(a)}] + E_{s,z \sim p(s,z)}[log \frac{p(z|s)}{p(z)}]
 = -E_{s,a,z \sim p(s,a,z)}[log \frac{\pi(a|s,z)}{p(a)}] + E_{a \sim p(a|s,z)}\big[E_{s,z \sim p(s,z)}[log \frac{p(z|s)}{p(z)}]\big] \\
 = E_{s,a,z \sim p(s,a,z)} [log \frac{p(z|s)}{p(z)} - log \frac{\pi(a|s,z)}{p(a)}] \ \ \ \ \text{where} \ \ \ p(a) = \int\int \pi(a|s,z) dsdz \\
\Rightarrow R(\pi) = E_{s,a,z \sim p(s,a,z)} [log \frac{p(z|s)}{p(z)} - log \frac{\pi(a|s,z)}{\pi(a)}]
$$

البته در نوتیشن مقاله فرض شده است با داشتن پالیسی و فرض دانستن داینامیک محیط می توان توزیع های مارجینال متغیر های تصادفی را از توزیع های حالت پایا با توجه به $p(\pi)$ بدست آورد و اکسپکتیشن ها به صورت خلاصه به شکل $E_\pi$ نوشته شده اند(مانند نوتیشن .$\tau \sim p(\pi)$ در حالت اکسپکتیشن روی ترجکتوری و مارجینالایز کردن توزیع های حالت پایا از آن)

سپس با توجه به غیر قابل محاسبه بودن $\pi(a)$ و $p(z|s)$ آنها با توزیع وردشی جایگزین شده اند و توزیع پرایور p(a) نیز یونیفورم در نظر گرفته شده و از تابع هدف بهینه سازی حذف شده است:

$$
 R(\pi) = E_{s,a,z \sim p(s,a,z)} [log \frac{p(z|s)}{p(z)} - log \frac{\pi(a|s,z)}{\pi(a)}] \\ 
 = E_{s,a,z \sim p(s,a,z)} [log p(z|s) + log q(z|s) - log q(z|s) - log p(z) - log \pi(a|s,z) + log \pi(a) +log p(a) - log p(a)] \\
= E_{s,a,z \sim p(s,a,z)} [log q(z|s) - log p(z) - log \pi(a|s,z) +log p(a)] + E_{s}[D_{KL}(p(z|s)||q(z|s))] + D_{KL}(\pi(a)||p(a)) \\
\geq E_{s,a,z \sim p(s,a,z)} [log q(z|s) - log p(z) - log \pi(a|s,z) +log p(a)] \ \ \ \ \ \text{where}  \ \ \ \ p(a) = \mathcal Unif \\
\Rightarrow  R(\pi) \geq E_{s,a,z \sim p(s,a,z)} [log q(z|s) - log p(z) - log \pi(a|s,z)]
$$

### الف-۱)
همانطور که در تمرین قبل دیدیم فرم سیاست در بهینه سازی در چهارچوب بیشینه انتروپی به شکل زیر است:

$$
\pi_{new} = argmin_{\pi \in \Pi} D_{KL}(\pi(.|s_t) || \frac{exp(Q^{old}(s_t))}{Z^{old}(s_t)})
\Rightarrow J(\pi_{\theta}) = E_{s_t \sim D} [ D_{KL}(\pi(.|s_t) || \frac{exp(Q^{old}(s_t))}{Z^{old}(s_t)}) ]\\
\theta_{new} = \theta + \alpha \nabla J(\pi_{\theta})
$$

در مقاله SAC ثابت شد که بهینه سازی بالا روی پالیسی باعث ارضای شرط policy improvement می شود. این بهینه سازیی در حالت غیر tabular که پالیسی یک شبکه عصبی است با محاسبه گرادیان تابع هدف معادله بالا صورت می گیرد. محاسبه این گرادیان می تواند به صورت آف پالیسی انجام شود و کافی است بعد از policy evaluation و محاسبه تابع ارزش پالیسی فعلی، از بچ سمپل گرفته شده و در دیتا پوینت های مختلف فاصله پالیسی با سافت مکس تابع ارزش کمینه شود.

در PPO اما هدف آن است که آپدیت پالیسی با دیتای فعلی به روش گرادیان سیاست به گونه ای کنترل شود که عملکرد سیاست کاهش نیابد و فرم بسته خاصی برای پالیسی ارائه نشده است که امکان بهبود آن را به صورت آف پالیسی میسر کند و در نتیجه مانند معمول روش های گرادیان سیاست آن پالیسی است. به طور خلاصه در مقاله فاز بهبود سیاست به شکل زیر نوشته شده است:

$$
\theta_{t+1} = arg \max_\theta E_{t} [L(s,a,\theta_t,\theta)]
$$

که آبجکتیو L توسط تکنیک های مختلف و با توجه به مقدار مزیت و چگالی احتمال اعمال در پالیسی فعلی آپدیت را کنترل کند. اما به هر حال گرچه از ایمپورتنس سمپلینگ و ... استفاده شده است اما همچنان آپدیت پالیسی $E_{s,a \sim \pi_t}$ روی دیتای جمع آوری شده توسط خود پالیسی انجام می شود که باعث می شود الگوریتم آن پالیسی باشد.

### الف-۲)
متا لرنینگ بر این اصل تاکید دارد که توزیع تسک در زمان متا-ترین باید با زمان متا-تست مشابه باشد. به همین دلیل متا آموزش یک ایجنت با دیتای آف پالیسی ساختار تسک در زمان متا-ترین را با ساختار داده ای که ایجنت در زمان متا-تست برای اداپت شدن به محیط جدید با اکتشاف آن-پالیسی از آن محیط به دست می آورد متفاوت می کند. به همین دلیل آموزش الگوریتم هایی مانند MAML باید آن پالیسی باشند و آف پالسی بودن با ذات مساله متا-یادگیری در تناقض است.

زمانی را پدید می آورد که به یادگیری توزیع ریوارد هایی خود تسک جدید نیز کمک بسیاری میکند. از طرفی مشاهده دیگر که در مقاله MAESN به آن اشاره شده است تفاوت اساسی رفتار اکتشاف و غیر قطعی بودن آن با رفتار بهره برداری و قطعی بودن آن است که باعث می شود در تسک های پیچیده تجمیع این ۲ رفتار در یک پالیسی با اکتشاف در سطح اعمال را سخت کند. مثلا در تسک یافتن هدف روی نیم دایره سیاست اکتشا کناسب حرکت روی نیم دایره است و سیاست بهره برداری مناسب حرکت از مرکز به سمت یک نقطه از محیط است. با توجه به این مشاهده و استفاده از دیدگاه احتمالاتی جدا سازی این ۲ عملیات یعنی استننتاج در مورد تسک و یادگیری بهره برداری تجربیات گذشته را جمع آوری میکند تا درر تسک های آینده برای استنتاج راجع به تسک استفاده شوند. در زمان متا تست تایم به دلایل مختلف مانند غیر قابل محاسبه بودن استنتاج روی کل توزیع باور روی تسکه ها، با سمپل گرفتن از این متغییر های context و استفاده از آن ها دبه عنوان ورودی پالیسی به عنوان یک متغیر تصمیم گیری برای یک اکتشاف پیوسته در زمان، توزیع باور پسین روی تسک جدید را به صورت تقریبا بهینه (با خواص یادگیریی عادی وظیفه اش توسط پالیسی انجام می شد و در صورت استفاده از دیتای آف پالیسی دچار مشکل می شد حال با دیتای آن پالیسی آموزش داده می شود، اما آموزش پالیسی که باتل نک سمپل افیشنسی ایجنت است حال می تواند با گرفتن متغیر های کانتکست با دیتای آف پالیسی توسط یک الگورتیم مانند SAC صورت گیرد.

همچنین کانتکست انکدر برای استنتاج راجع به تسک باید داینامیک MDP متناظر را شناسایی کند و این مدل قاعدتا نباید ربطی به ترتیب داده هایی که به آن داده می شود داشته باشد و باید در سطح ترنزیشن ها و نه ترتیب مشاهده انها باشد زیرا این ترتیب بیشتر به پالیسی وابسته است نه داینامیک محیط. به همین منظور تسک انکدر به صورت یک مدل permutation invariant ضرب فاکتور های گاوسی در نظر گرفته شده است که ورودی آن ترنزیشن ها و احتمال یک ترجکتوری ضرب فاکتور های ترنزیشن های ان بدون توجه به ترتیب وقوع آنهاست.

### ب)
 در مسائل تک وظیفه اطلاعات ناشناخته برای ایجنت داینامیک محیط است و با اکتشاف در محیط و شناختن توابع انتقال و پاداش به اندازه کافی می تواند قطعیت خود راجع به محیط را افزایش داده و از اطلاعات بدست امده با اطمینان بهره برداری کند. اما در شرایط چند وظیفه ای یک متغیر نهان در مشاهدات عامل از محیط وجود دارد و آن وجود عدم قطعیت در تشخیص خود محیط و امکان تغییر داینامیک محیط در لحظات متفاوت است. بدین معنی که ممکن است مشخصات محیط عامل تغییر کند و دیگر اطلاعات کشف شده عامل در محیط های قبلی در محیط فعلی صدق نکند اما عامل از این تغییر خبر نداشته باشد. به همین منظور اکتشاف باید از سطح شناسایی محیط فراتر رفته و به کاهش عدم قطعیت در مورد توزیع محیط های ممکن نیز بپردازد. در نتیجه اکتشاف از سطح عمل که اطلاعات انتقال و پاداش را به دست می آورد کافی نیست و باید اکتشافی پیوسته و مرتبط با اکتشافات قبلی به منظور کسب اطمینان از در نظر گرفتن تغییر وظیفه نیز انجام شود.


ایراد استفاده از خطای بیش بینی مدل محیط نیز همین در بر نگرفتن این متغییر نهان است. زیرا این خطا باعث میشود که ترنزیشن هایی پاداش داخلی بالاییی دریافت کنند که مدل ما از داینامیک آنها عدم قطعیت دارد و با انجام آنها شناخت ما از محیط بهبود می یابد، اما با تغیییر متغیر نهان محیط و تغیییر احتمالی در داینامیک محیط عامل، این مدل دیگر شناخت ما از محیط فعلی را نمایندگی نمی کند اما عامل همچنان به آن اتکا دارد. به عنوان مثال ترنزیشنی که در محیط قبلی به خوبی درک شده بود ریوارد کمی خواهد داشت اما ممکن است در محیط جدید این ترنزیشن احتمال کمی داشته باشد و ما اطلاعات مناسبی از تاباع پاداش در این قسمت نداشته باشیم اما چون ما همچنان از مدل محیط قبلی به عنوان پاداش داخلی استفاده می کنیم، پاداش این ترنزیشن همچنان کم است و عامل اکتشاف مناسبی در این نقاط محیط جدید ندارد. با ایین اطلاعات تصحیح نشده در محیط جدید بهره برداری از پاداش ها نیز ضعیف خواهد بود.

واضحا وقتی امبدینگی از تسک ها نیز در مدل محیط لحاظ شود امکان در نظر گرفتن این متغیر نهان به وجود می آید و در صورت کیفیت مناسب امبدینگ هم تغییرات محیط در اکتشاف لحاظ می شود و هم از اکتشافات عامل در محیط های قبلی که ساختار مشابه محیط فعلی دارند نیز می توان استفاده مجدد کرد و تعمیم پذیری و کیفیت و سرعت اکتشاف را بالا برد که از اهداف یادگیری چند وظیفه ای و متا یادگیری است. البته برای این اتفاق نیاز به اطلاعات مناسب از وظیفه در دست و امبدینگ منناسب ان می باشد که گاها در زمان متا-تست ممکن نیست.


### ج-۱)

استفاده از توزیع باور روی تسک ها مشکلات محاسباتی دارد. هم محاسبه و نگه داری توضیع پسین و مارجینالایزز کردن آن روی همه تسک ها از لحاظ محاسباتی با افزایش پیچیدگی تسک ها به زودی غیر ممکن می شود، هم با فرض محاسبه این توزیع پسین استفاده مستقیم از آن در بهینه سازی سیاست غیر ممکن است. زیرا برای استفاده از این توزیع ممعمولا باید استیت اسپیس را با حالت باور ترکیب کرد که حالت باور توضیع پسین است و محاسبات یادگیری سیاست در این فضای حالت جدید انجام گیرد تا عدم قطعیت نیز در نظر گرفته شود اما بهیینه سازی در این فضای حالت جدید برای تسک های ساده نیز مشکل است. به همین دلیل در یادگیری تقویتی از روشی به نام تامپسون سملنیگ یا نمونه گیری از توزیع پسین استفاده می شود که هر چند استپ (یا اپیزود) یک نمونه از توزیع پسی گرفته می شود و اکتشاف و بهره برداری و بهینه سازی سیاست با فرض عمل کردن در محیط سمپل شده انجام می شود. در نهایت با توجه به عملکرد و آبجکتیو عامل، باور ایجنت نسبت به محیطی که فرض بودن در آن داشت تست می شد اصلاح می شود( این کار مثلا با عبور گرادیان آبجکتیو پالیسی با reparametrization trick از نمونه ی گرفته شده از توزیع پسین و اصلاح پارامتر های مدل در نظر گرفته شده برای توزیع تسک ها (مانند فاکتور های گاوسی در pearl) صوررت می گیرد). به این ترتیب نیازیی به یادگیریی سیاست در فضای حالت بزرگ ترکیب شده با باورر نیست و به جای اصلاح عدمم قطعیت به صورت سیستماتیک برای هر باور در هر استیت، در همه استیت ها این کار با نمونه گیری و مانند جایگزینی محاسبه انتگرال اکسپکتیشن با سمپل گرفتن انجام می شود و اصلاح باور ها به صورت میانگین با سمپل گرفتن در استیت های مختلف صورت می گیرد. این تکنیک bayes-optimal نیست اما در تئوری نشان داده شده است در عملکرد حدی رگرت آن با روش های اپتیمال مشابه است و در عین حال از لحاظ محاسباتی نیز بهینه است. همچنین در یادگیری تقویتی این تکنیک مزایایی مانند اکتشاف عمیق و با پیوستگی زمانی نیز دارد زیرا چند اکشن پشت سر هم برای تست کردن یک فرضیه انتخاب می شوند و عملکرد آنها به تنهایی مانند روش های خوش بینی در مواجهه با عدمم قطعیت یا دیگر روش های اکتشاف بررسی نمی شود. 



### ج-۲)
ابتدا آبجکتیو کلی را می نویسیم: با داشتن یک پالیسی که روی تسک-دسکریپشن کاندیشن شده است توزیع ترجکتوری های بدست آمده با اجرای این پالیسی در داینامیک نامشخص محیط ورودی را به نسبت پالیسی و احتمالا تسک انکدر که z از آن گرفته می شود (یا اگر پالیسی روی کل توزیع تسک کاندیشن شده باشد خود انکدر) اپتیمایز میکنیم:

$$
J(\pi) = E_{z \sim p(z), \tau \sim (\pi|z)} [p(\tau)]
$$

به این منظور باید از یک مدل به شکل زیر استفاده کرد:

اما همانطور که دیدیم محاسبه $p(\tau)$ و مارجینالایز کردن آن نسبت به توزیع تسک ها z از لحاظ محاسباتی سخت است. به همین دلیل اکسپکتیشن $p(\tau|z)$ را روی توزیع وردشی پسین تسک ها می گیریم. به کمک باند پایین نوشته شده در صورت سوال آببجکتیو به شکل زیر در می آید:

$$
J(\theta, \phi) = E_{z \sim p(z), \tau \sim (\pi_\theta|z)} \big[ E_{q_\phi(z|\tau)}[p_{\psi}(\tau|z)] - D_{KL}(q_\phi(z|\tau)||p_\theta(z)) \big]
$$

به این ترتیب ساختار کلی مدل مشخص می شود: فلش $D_{i}^{tr}->z_i$ یک کانتکست انکدر است که با ورودی گرفتن ترجکتوری امبدینگ کانتکست و یا همان حالت باور عامل نسبت به محیط را خروجی می دهد. باند elbo با ریپارامتریزیشن بهینه می شود و همچنین با دادن z به پالیسی به صورت end-to-end با هم ترین می شوند. ترم ریکانستراکشن که به صورت $E_{q_\phi(z|\tau)}[p(\tau|z)]$ ظاهر شده است با استفاده از یک دیکدر مانند $P(s,s^\prime,z)$ در شکل بالا به صورت اند تو اند با انکدر ترین می شود. در نتیجه لاس به صورت جمع لاس پالیسی و لاس دیکدر جمع می شود و سه شبکه با هم آموزش داده می شوند. برای آموزش بهتر انکدر و دیکدر می توان به صورت همزمان و در طی مسیر که قدم به قدم از سیاست ترنزیشن ها سمپل گرفته می شوند، هر بار توزیع پسین $q_\phi(z|\tau_{0:t})$ به حای توزیع پیشین p(z) جایگزین شود و ترنزیشن جدید $(s_t,a_t,s_{t+1})$ به انکدر ورودی داده شود تا با کانتکست قبلی و توزیع پرایور جدید هر بار یک ترم لاس برای کانتکست از شروع تا حال به لاس انکدر اضافه شود. در دیکدر نیز می توان برای لاس هر مرحله زمانی، ترنزیشن آن مرحله به عنوان تارگت در نظر گرفته شود تا به نوعی ریکانستراکشن ارور در دیکدر همان لاس Dynamics Prediction Error ذکر شده در قسمت قبل سوال باشد. به این ترتیب دیتا مناسب نیز برای آموزش دیکدر در دسترس خواهد بود.

این روش همانطور که پیداست یک روش black-box متا لرنینگ است و می توان از شبکه های RL2 برای سیاست و شبکه های عصبی مبتنی بر حافظه برای انکدر استفاده کرد. بدیین ترتیب همه ی پارامتر های تتا و فای (و حتی سای دیکدر) متا پارامتر هستند و در زمان متا تست نیز اداپتیشن با یک فروارد پس روی انکدر (برای هر ترنزیشن)، بدست آوردن پارامتر های مختص تسک که همان توزیع حالت باور z هستند و سپس اجرای متا پالیسی ای که یاد گرفته است چگونه با گرفتن مشخصات مناسب تسک پاداش را بهینه کند اداپتیشن انجام می شود. توجه شود که دیکدر در زمان متا اموزش فقط برای خطای بازسازی انکدر آموزش داده می شود و در زمان اداپتیشن بی استفاده است.

نکته دیگر این است که گفتیم خروجی کانتکست انکدر (z) نمایانگر عدم قطعیت در تسک ها و توزیع باور روی تسک ها است. برای بهینه سازی آسان تر باند پاییین انکدر دیکدر شبکه انکدر پارامتر های آماری توزیع را خروجی می دهد (مانند میانگین و واریانس یک توزیع گاوسی) و پالیسی روی انها کاندیشند می شود که مشکل غیر قابل محاسباتی بودن یادگیری پالیسی در فضای حالت ترکیبیی با حالت باور را حل میکند و همچنین به روز کردن توزیع پسین نیز با rreparametrization trick آسان می شود. 

### د-۱) 
در دیدگاه MAML بهینه سازی داخلی یک تابع تصادفی بود که گرادیان آبجکتیو روی توزیع ترجکتوری های پالیسی پس از آپدیت را به پارامتر های متا پالیسی منتقل میکرد. بدیین ترتیب این گرادیان از فرآیند بهینه سازی داخلی عبور می کرد و باعث تخصیص اعتبار به این فرآیند و درر نتیجه بهبود نقطه اولیه پارامتر هایی تولید کننده ترجکتوری هایی اکسپلوریشن با نگاه به پاداش دریافتی پس از بهینه سازی داخلی میشد. اما روش جدید تابع بهینه سازی داخلی را به چشم یک تابع قطعی روی تعدادی ترجکتورری سمپل گرفته شده از تسک نگاه می کند و اکسپکتیشن روی این ترجکتوری ها خارح از تابع آپدیت است. این دیدگاه به فرآیند بهینه سازی اداپتیشن داخلی اعتباری برای پاداش های نهایی تخصیص نمی دهد و از ساختار واقعی فرآیند یادگیریی بهره نمی برد و بهینه سازی داخلی را جزئی از داینامیک نامشخص محیط می داند که نمی توان از طریق عبور گرادیان از آن مقادیر اولیه مناسب متا پارامتر ها  را بهبود داد. اما به هر حال با رساندن گرادیان پاداش های نهایی به متا پارامتر ها سعی در بهبود توزیع ترجکتوری های سمپل گرفته شده در حین اداپتیشن (ترجکتوری های ورودی بهینه سازی داخلی) که حاصل از اجرای متا پالیسی یادگرفه شده هستند می شود. اما روش اولیه MAML علاوه بر این مساله به دلیل تخصیص اعتبار به فرایند بهینه سازی داخلی و در نظر گرفتن تاثیر علی توزیع ترجکتوری های متا پالیسی زمان اکسپلوریشن و پالسی اداپت شده علاوه بر بهبود مسیر های اکسپلوریشن نسبت به ریوارد، این توزیع اکسپلوریشن را برای داشتن قدم های سریع تر تابع بهینه سازی داخلی با نگاه به ریوارد توزیع ترجکتوری های پس از اداپتیشن نیز اپتیمایز می کند که باعث متا یادگیری اساسی تر و خواص اکسپلوریشن سریع تر و قوی تر می شود. در نتیجه در روش اصلی MAML توجه به نقطه اولیه مناسب تری برای اکتشاف می رسد. 

### د-۲,۴)
برای خلوت تر شدن عبارات اکسپکتیشن روی توزیع تسک ها را نمی نویسیم و محاسبات را روی آبجکتیو برای تسک i ام در نظر میگیریم:

طبق معادلات ۱۵ تا ۱۷ مقاله:

$$
\nabla_{\theta} J^{MAML}_i(\theta) = E_{\tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} [\nabla_{\theta^\prime} log P_i(\tau^\prime | \theta^\prime) R(\tau^\prime) \nabla_\theta \theta^\prime]
$$

که با توجه به اینکه بهینه سازی داخلی یک گام گرادیان نزولی روی اکسپکتیشن پاداش های متا پالیسی است داریم:

$$
\nabla_\theta \theta^\prime = \nabla_\theta U(\theta, T_i) = \nabla_\theta (\theta + \alpha \nabla_\theta E_{\tau \sim P_i(\tau | \theta)} [R(\tau)] \\
= I + \nabla^2_{\theta} E_{\tau \sim P_i(\tau | \theta)} [R(\tau)]
$$

که خود هشن را هم طبق معادلات ۲۰ تا ۲۴ مقاله می توان به شکل زیر بازنویسی کرد: (جزئیات معادله بسیار طولانی بود اما واضح است چگونه به این معادله رسیده شده. ابتدا یکی از گرادیان ها را به داخل اکسپکتیشن برده است و با استفاده از قانون: گرادیان اکسپکتیشن = اکسپکتیشن گرادیان لاگ، در مرحله اول هشن اکسپکتیشن پاداش به گرادیان اکسپکتیشن لاگ پالیسی ضربدر پاداش تبدیل شده است. در مرحله دوم با وارد کردن گرادیان دوم با همان قانون: گرادیان اکسپکتیشن = اکسپکتیشن گرادیان لاگ، و با توجه به اینکه حال داخل اکسپکتیشن ضرب ۲ عبارت است، گرادیان لاگ عبارت داخل به ۲ ترم زیر تبدیل شده است)

$$
\nabla^2_{\theta} E_{\tau \sim P_i(\tau | \theta)} [R(\tau)] = E_{\tau \sim P_i(\tau | \theta)} \big[R(\tau)[\nabla^2_\theta log \pi_\theta(\tau) + \nabla_\theta log \pi_\theta(\tau) ^T \nabla_\theta log \pi_\theta(\tau)]\big]
$$

با جاگذاری ۲ معادله بالا در گرادیان آبجکتیو MAML داریم:

$$
\nabla_{\theta} J^{MAML}_i(\theta) = E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} \big[ \nabla_{\theta^\prime} log P_i(\tau^\prime | \theta^\prime)(R(\tau^\prime)(I+\alpha R(\tau)\nabla^2_\theta log \pi_\theta(\tau))) + \nabla_\theta log P_i(\tau|\theta)(\alpha \nabla_{\theta^\prime}log P_i(\tau^\prime|\theta^\prime)R(\tau^\prime)R(\tau)\nabla_{\theta}log P_i(\tau|\theta))^T\big] \\
= E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} [ C_1\nabla_{\theta^\prime} log P_i(\tau^\prime | \theta^\prime) + C_2 \nabla_\theta log P_i(\tau|\theta)] \\
E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} [ C_1\nabla_{\theta^\prime} log \pi_{\theta^\prime}(\tau^\prime) + C_2 \nabla_\theta log \pi_{\theta}(\tau)]
$$

ترم اول گرادیان مسئول بهینه سازی مقدار اولیه متا پالیسی با اعمال یک تبدیل خطی گرادیان پالیسی پس از اداپتیشن به پارامتر ها اولیه با وزن پاداش نهایی است. این ترم که در هر ۲ فرمول یکسان است متا پارامتر ها را با نگاه به پاداش پس از اداپتیشن جا به جا میکند. با اینکه ارتباط پارامتر ها پیش و پپس از بهینه سازی داخلی را در نظر می گیرد اما گرادیان نسبت به توزیع پالیسی پس از آپدیت است که تبدیل یافته است و مستقیما توزیع اولیه را بهینه نمی کند. ترم دوم مسئول تنظیم توزیع ت رجکتوری های پیش از اداپتیشن حاصل از متا پالیسی اولیه است که وزن این آپدیت با پاداش پس از اداپتیشن و ارتباط علی گرادیان های پیش از اداپت و پس از آن توسط C2 مشخص می شود. در نظر نگرفتن این ارتباط علی همانطور که در روش دوم دیده خواهد شد، ممکن است گاها باعث شود که تغییر در متا پارامتر های در خلاف جهت گرادیانی باشد که به نسبت پالیسی پس از اداپتیشن گرفته شده اند (ترم C1) و عملکرد کلی را کاهش دهد. اما در نظر گرفتن این ارتباط علی در آپدیت توزیع پیش از اداپتیشن باعث هم جهت شدن گرادیان توزیع هایی پییش و پس از اداپت به منظور افزایش پاداش نهایی خواهد شد.

می توان آبجکتیو روش دوم را نیز به همین دیدگاه آبجکتیو روی توزیع پیش و پس از اداپتیشن تقسیم کرد. با توجه به عبارت ۲۶ مقاله و جاگذاری عبارات ۲۷ و ۲۸ در آن به عبارت ۳۴ و ۳۵  میرسیم و فرم نهایی آبجکتیو را به آبجکتیو روی توزیع پیش و پس از اداپتیشن تقسیم می کنیم:

$$
(26): \nabla_\theta J^{E-MAML} (\theta) = E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} \big[ \nabla_{\theta^\prime} log P_i(\tau^\prime | \theta^\prime)(R(\tau^\prime)(I+\alpha R(\tau)\nabla^2_\theta log \pi_\theta(\tau))) + \nabla_\theta log P_i(\tau|\theta)R(\tau^\prime)\big] \\
= E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} [ C_1\nabla_{\theta^\prime} log P_i(\tau^\prime | \theta^\prime) + C_2 \nabla_\theta log P_i(\tau|\theta)] \\
E_{\tau \sim P_i(\tau | \theta), \tau^\prime \sim P_i(\tau^\prime | \theta^\prime)} [ C_1\nabla_{\theta^\prime} log \pi_{\theta^\prime}(\tau^\prime) + C_2 \nabla_\theta log \pi_{\theta}(\tau)]
$$

که در اینجا C1 همان C1 روش MAML است و همانطور که گفته شد C2 به ارتباط توزیع پیش و پس از اداپتیشن و فرآیند بهینه سازی داخلی اعتباری برای دریافت پاداش تخصیص نمی دهد که باعث می شود گرادیان های عبوری از مسیر پیش و پس از اداپتیشن امکان تداخل با یکدیگر را داشته و متا پارامتر های اولیه اکسپپلوریشن ضعیف تری داشته باشند.



### د-۳)
همانطور که در گراف های محاسباتی در مقاله دیده می شود تفاوت ۲ روش در قسمت بهینه سازی داخلی و رسیدن از $\pi(\theta)$ به $\pi(\theta^\prime)$ است و پس از آن توزیع $\tau^\prime$ و توزیع پاداش های ترجکتوری های سیاست پس از اداپتیشن یکسان است. به همین دلیل مسیر گرادیان از آبجکتیو تا $pi(\theta^\prime)$ یکسان است. انتقال این گرادیان به تتای اولیه از طریق تبدیل تتا پرایم به تتا است که گرچه مسیرهای گراف محاسباتی ۲ روش در این قسمت متفاوت است اما تابع تبدیل تتا به تتا پرایم همان بهینه سازی داخلی گرادیان نزولی است و در هر ۲ روش ضریب گرادیان زنجیره ای مشتق آبجکتیو نسبت به تتا با ارزیابی در تتا پرایم همان ضریب مشتق گرادیان تتا پرایم نسبت به تتا است. 


### د-۵)

در قسمت های قبل بارها ذکر شد که این ترم بهینه سازی پارامتر های اولیه را نه فقط به سادگی با در نظر گرفتن پاداش ها به عنوان وزن، بلکه با در نظر گرفتن اثر علی پالیسی اولیه روی پالیسی پس از اداپت و سپس پاداش آن پالیسی در نظر می گیرد. این امر با وزن دادن پاداش با ضرب گرادیان های پیش و پس از اداپت انجام می شود.این ضرب داخلی تعبیری مانند کواریانس گرادیان پالیسی با پاداش های اکسپلوریشن(گرادیان پیش از اداپت) و پالیسی اداپت شده با پاداش اکسپلویتیشن (گرادیان ترم post) دارد و باعث می شود که آپدیت ساده انگارانه اولیه توزیع پیش از اداپت ($J_{pre}^{E-MAML}$) با ضرب کردن این ضرب داخلی، با گرادیان پاداش نسبت به توزیع پس از اداپت ($J_{post}$) تداخل نداشته باشد (رابطه ۳۸ مقاله) و به صورت کاملا بهینه توزیع اولیه تنظیم شود.

### د-۶)
مشکل ذکر شده واضحا در آبجکتیو E-MAML وجود دارد زیرا گرادیان ترجکتوری های اکسپلوریشن فقط با پاداش دریافتی ترجکتوری بهره وزن داده شده است و همانطور که در DREAM گفته شد، وقتی پالیسی بهره برداری در مراحل اولیه که اطلاعات اکتشاف کم است، ضعیف است پاداش دریافتی آن نیز کم است و در نتیجه اپدیت پالیسی اکتشاف E-MAML که فقط به این پاداش وابسته است مناسب نخواهد بود و همچنان اکتشاف و بهره در یک لوپ عدم پیشرفت قرار میگردند

در آبجکتیو MAML امابا توجه به اینکه گرادیان خود پالیسی اکتشاف نیز با پاداش ترجکتوری اکتشاف در پاداش بهره و گرادیان آن ضرب شده است، حتی اگر پالیسی بهره برداری ضعیف باشد اطلاعات بدست آمده از ترجکتوری اکسپلوریشن به خود پارامتر های اکتشاف منتقل می شود(البته تا حدودی، اگر پالیسی بهره زیاد گمراه نباشد) و همچنان پیشرفتی در پالیسی اکتشاف خواهیم داشت تا شاید در نهایت بهینهه سازی end-to-enddاز لوپ خارج شود.


### ه-۱)
روش های معمول بلک باکس مانند RL2 تابع زیان J را به صورت end-to-end و هم زمان برای هر ۲ پالیسی اکتشاف و بهره با توجه به پاداش دریافتی بهره برداری بهینهه سازی می کنند. در اصل در این روش ۲ سیاست پارامتر های کاملا مشترک دارند که یک پالیسیی رییکارنت است که علاوه بر استیت هیستوری را نیز ورودی میگیرد. ابتدا پالیسی که با متا پارامتر ها مقدار دهی شده است یک اپیزود اکسپلوریشن ایجاد میکند، سپس استیت شبکه بازگشتی به عنوان کانتکست و متغیر هایی مختص تسک بعد از اداپتیشن دوباره به خود پالیسی داده می شود و از اینجا به بعد به عنوان پالیسی بهره برداریی تعدادی اپیززود با توجه به تجربه ایجاد شده در اکشتاف تسک ایجاد می کند. در نهایت اعتبار تابع زیان به هر دو مرحله اکتشاف و بهره برداری از طریق بازگشت گرادیان از مسیر شبکه عصبی بازگشتی انجام می گیرد. این روش علاوه بر پیچیدگی بهینه سازی اش، می تواند در یک لوکال اپتیمم نیز گیر بیوفتد که دلیل آن فرایند بهینه سازی در هم آمیختهه ۲ سیاست است مخصوصا که درر روش RL2 دو سیاست یکسان هستند. زیرا بهینه سازی سیاست اکتشاف از طریق بازگشت گرادیان عبوری از سیاست بهره است. اگر سیاست بهره برداریی در حل مساله ضعیف باشد گرادیان های سیاست اکتشاف تاثیر مثبت نخواهند داشت، همچنین ضعیف بودن اطلاعات سیاست اکتشاف در مرحله بعد فایده ای برای پیشرفت سیاست بهره برداری ندارد. از آنجا که هر ۲ سیاست در شروع آموزش اپتیمال نیستند این اتفاق می تواند آموزش را در یییک لوپ و لوکال اپتیمم نگه دارد.

روش DREAM بهینه سازی این ۲ سییاست را با فرض دسترسی داشتن به اطلاعات تسک در زمان آموزش انجام می دهد. به این صورت که سعی می شود از مشخصات تسک اطلاعات مفید و متمایز کننده تسک استخراج شود و از این اطلاعات برای آموزش سیاست بهره برداری استفاده شود ( و در نتیجه به صورت همزمان در این فرآیند استخراج کننده اطلاعات نیز آموزش داده می شود). از آن طرف یک سیاست اکتشاف به نحوی آموزش داده می شود که توانایی بازتولید این اطلاعات موثق و مفید را پیدا کند. به این صورت یادگیریی سیاست بهره ربطی به کیفیت اطلاعات سیاست اکتشاف ندارد که مکنن است اپتیمال نباشند و همچنین سیاست اکتشاف گرادیانش را از خواص تسک می گیرد نه فیدبک سیاست بهره که ممکن است بهینه نباشد. 


### ه-۲)
در این تابع هزینه یک بهینه سازی مقید داریم که هدف آن کمینه کردن اطلاعات موجود در امبدینگ z از مشخصات وظیفه است در شرایطی که سیاست بهره برداری یادگرفته شده با ورودی گرفتن این امبدینگ که خروجی انکدر F است، عملکرد بهینه ای در جمع آوری پاداش از محیط داشته باشد (تابع ارزش آن با تابع ارزش بهینه مساوی شود). این تابع هدف سعی میکند هرگونه اطلاعات غیر مفید برای بهینه عمل کردن عامل در ممحیط را از امبدینگ تسک خارج کند تا بعدا سیاست اکتشاف یاد گرفته شده روی این امبدینگ نیز فقط استخراج اطلاعات مفید را یاد بگیرد. دلیل اضافه کردن این باتل نک اطلاعات همین تشویق اکتشاف هدف مند و بهینه است. همچنین از مزایای دیگر این تابع هدف بهینه سازیی سییاست بهره برداری روی اطلاعات مفیدی است که وابستگی ای به کیفیت سیاست اکتشاف ندارند.


### ه-۳)
همانطور که گفته شد بعد از آموزش انکدری که اطلاعات مفید برای حل بهینه تسک توسط سیاست بهره برداری را از اطلاعات تسک استخراج می کند، می توانیم بهینه سازی سیاست اکتشاف را با این هدف انجام دهیم که ترجکتوری هایی ایجاد کند که اطلاعات مشابه را تولید کنند. این کار را می توان با انتخاب اطلاعات متقابل ترجکتورری هایی تولید شده سیاست اکتشاف و این امبدینگ به عنوان هدف و بییشینه سازی آن انجام داد:

$$
I(\tau^{exp};z) = H(z) - H(z|\tau^{exp}) = H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}}[log p(z|\tau^{exp})]] = H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}}[log q(z|\tau^{exp})]] + D_{KL}(p(z|\tau^{exp})||q(z|\tau^{exp})) \geq H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}}[log q(z|\tau^{exp})]]
$$

از آنجایی که توزیع $p(z|\tau^{exp})$ قابل محاسبه نیست آن را با یک توزیع وردشی آن جایگزینن کرده و باند پایین آن را بهینه میکنیم. این توزییع وردشی در زمان متا تست که به مشخصات تسک دسترسی نداریم و سمپل گرفتن از F ممکن نیست نقش دیکدر را را دارد که از تجربه اکتشاف امبدینگ تسک را تولید می کند برای سیاست بهره برداری

$$
H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}}[log q(z|\tau^{exp})]] = H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}} [log q(z|\tau^{exp}_{:0} - log q(z|\tau^{exp}_{:0} + log q(z|\tau^{exp}_{:1} - log q(z|\tau^{exp}_{:1} + ... + log q(z|\tau^{exp}_{:T}] = H(z) + E_{\mu}[E_{z \sim F(z|\mu), \tau^{exp} \sim \pi^{exp}}[log q(z|s_0) + \sum_{t=0}^{T-1} log \frac{ q(z|\tau^{exp}_{:t+1})}{ q(z|\tau^{exp}_{:t})}]
$$

۲ ترم اول در بهینه سازی سیاست بی تاثیرند. ترم سوم نیز به صورت جمع یک سری ترم در مسیر تولید شده توسط سیاست را دارد که می توان به آن به چشم بهینه سازیی یک مساله یادگیری تقویتی با پاداش $log \frac{ q(z|\tau^{exp}_{:t+1})}{ q(z|\tau^{exp}_{:t})}$ در هر مرحله t از مسیر نگاه کرد. این پاداش انجام اعمالی را در s_t تشویق می کند که اطلاعات مرتبط با تسکی که استخراج میکنند ($q(z|\tau^{exp}_{:t+1})$) چیزی برای اضافه کردن به اطلاعات موجود فعلی در ترجکتوری ایجاد شده ($q(z|\tau^{exp}_{:t})$) داشته باشد. همچنین می توان یک ثابت -c به این پاداش ها افزود تا با پاداش منفی تعداد مراحل اکتشاف نیز مجازات داشته باشند تا سرعت اکتشاف نیز تشویق گردد .

### ه-۴)
می توان به صورت unsupervised و مشابه ایده مقاله های skill-discovery یک pre training بدون نظارت برای ایجنت انجام داد. این پیش آموزش می تواند روی محیط ها مختلف یادگیری تقویتی باشد. درر این مرحله سعی می شود یک توزیع از پارامتر های مختلف داینامیک محیط تمرینی به همراه یک انکدر به صورت موازی به نحوی آموزش داده شوند که توزیع محیط ها شامل محیط های پراکنده با آبجکتیو ها و سیاست های بهینه پراکنده باشد و همچنین انکدر آموز داده شود که از تجربیات محدود عامل در محیط ها سریعا یک امبدینگ مناسب استخراج کند.
سپس در زمان متا ترین به جای استفاده از $F(z|\mu)$ از انکدر پیش آموزش داده شده استفاده شود تا همزمان با آموزش سیاست بهره برداری در قسمت اول DREAM به محیط جدید اداپت شود. اداپتیشن انکدر می تواند به صورت hierarchial و با ترکیب انکدینگ محیط هایی که در پیش آموزش توزیع آنها بهینه شد نیز بدست آید تا در استخراح امبدینگ مناسب از محیط سریعتر عمل کرده و سیاست بهره برداری نیز سریعتر به اطلاعات مفید دسترسی پیدا کند.
